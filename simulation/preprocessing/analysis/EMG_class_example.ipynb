```python
import numpy as np
import pandas as pd

start_flag_external = 670
end_flag_external = 900

# 計算數據點的數量
external_time = np.linspace(0, 1000, num=1001)
external_values = np.linspace(0, 10, num=1001)  # 模擬一條直線
external_series = pd.Series(data=external_values, index=external_time)
additional_data_filtered = external_series[(external_series.index >= start_flag_external) & 
                                          (external_series.index <= end_flag_external)]

additional_data = {
    'data': additional_data_filtered,          # 傳入篩選後的 RHeel_Y 數據
    'label': 'RY',                        # 圖例標籤
}
```

```python
emg_data.plot_emg(
    muscle_name='LTA', 
    start_flag=start, 
    end_flag=end, 
    show_raw=True,
    show_processed=False, 
    show_envelope_raw=False, 
    show_normalized=True,
    additional_start_flag=start_flag_external,
    additional_end_flag=end_flag_external,
    additional_data=additional_data
)
```
